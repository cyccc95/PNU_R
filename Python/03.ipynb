```python
import pandas as pd
```

```python
df = pd.read_csv("../03/03_암발생자수.csv", encoding="cp949")
df
```

```python
df.columns = ['암종별','성별','연령별','2019','2019.1']
df = df[(df['암종별'] == '모든 암(C00-C96)') & (df['연령별'] != '계') & 
(df['연령별'] != "연령미상")]
df
```

```python
import numpy as np
```

```python
df['연령대'] = np.where(df['연령별'].isin(df['연령별'][0:8]), "30대이하",
                np.where(df['연령별'].isin(df['연령별'][8:12]), "40~50대",
                np.where(df['연령별'].isin(df['연령별'][12:16]), "60~70대", "80대이상")))
df
```

```python
df['2019'] = df['2019'].astype('int')
df.info()
df2 = df.groupby(['연령대','성별']).sum()['2019']
df2
```

```python
df2 = df2.reset_index()
df2
```
